Coingecko API returns unix timestamps by default, in particular miliseconds since epoch, need to convert this and provide with a pandas datafram for furhter processes.
##### Examples
- https://stackoverflow.com/questions/19231871/convert-unix-time-to-readable-date-in-pandas-dataframe

```python
df = pd.DataFrame(hourly_btc_chart['prices'])
df.columns = ["timestamp", "price"]
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
df['currency'] = 'GBP'
df
```

```python
df.plot(x="timestamp", y="price")
```
