## A bit of TA...
Because we don't want to request the data too much, just do a bit here.

```python
hist
```

```python
data = hist.drop(['Dividends', 'Stock Splits'], axis=1)
data
```

```python
data.columns = ["open", "high", "low", "close", "volume"]
```

```python
data
```

```python
import pandas as pd
from finta import TA
import mplfinance as mpf
```

```python
# calc bol band
bbands = TA.BBANDS(data, 100)

# cherry pick what to show on the chart
bands_plot = pd.concat([bbands.BB_UPPER, bbands.BB_LOWER], axis=1)

apd = mpf.make_addplot(bands_plot.tail(300))

mpf.plot(data.tail(300), type='candle', style='charles',
        title='MSFT BBANDS(100)',
        ylabel='Price (USD)',
        ylabel_lower='Volume',
        volume=True,
        figscale=1.5,
        addplot=apd
        )
```
